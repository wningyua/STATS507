Your function should ignore spaces and capitalization in assessing whether or not a string is a palindrome, so that "tacocat" and "T A C O cat" are both considered palindromes.

```python
# YOUR CODE HERE
def is_palindrome(string):
    string = string.lower()# change to lower case
    string = string.replace(" ", "")# remove white space.
    len2 = len(string)//2
    for i in range(0, len2):
        if string[i]==string[-(i+1)]:continue
        else: return False
    return True
```

3. <b>(2 points)</b> Write a function called `binary_search` that takes two arguments, a list of integers `t`(which is guaranteed to be sorted in ascending order) and an integer `elmt`, and returns True if `elmt` appears in list `t` and False otherwise. The binary search algorithm works as follows: 

    Search a sorted array by repeatedly dividing the search interval in half. Begin with an interval covering the whole array. If the value of the search key is less than the item in the middle of the interval, narrow the interval to the lower half. Otherwise narrow it to the upper half. Repeatedly check until the value is found or the interval is empty. (Source: https://www.geeksforgeeks.org/binary-search/)
    
    __Note:__ No credit will be given if binary search is not implemented. A simple for loop checking that the element is in the list will not qualify as a binary search.

```python
def binary_search(t, elmt):
    mid=t[len(t)//2]
    # print(mid)
    if elmt == mid:
        return True
    elif elmt < mid: # go to left part
        t = t[:len(t)//2]
        if t == []:
            return False
        else:
            return binary_search(t, elmt) 
    else: # go to right part
        t = t[len(t)//2+1:]
        # print(t)
        if t == []:
            return False
        else:
            return binary_search(t, elmt)
        
binary_search(list(range(1,100)), 1900)
```

```python
# YOUR CODE HERE
def char_hist(string):
    string = string.lower()
    d = dict()
    for c in string:
        if c not in d:
            d[c] = 1
        else:
            d[c] = d[c]+1
    return d
```

```python
# YOUR CODE HERE
def bigram_hist(string):
    d = dict()
    string = string.lower()
    string = string.replace(" ", "")
    for i in range(len(string)-1):
        pair = tuple(string[i:(i+2)])
        if pair not in d:
            d[pair] = 1
        else:
            d[pair] = d[pair]+1
                
    return d

bigram_hist(' Ss\tsS SS ')
```

```python
# YOUR CODE HERE
def check_valid_mx(x):
    # print(type(x))
    if type(x) != tuple :
        return False
    for n in x:
        if type(n)!=tuple:
            return False
        else:
            for k in n:
                # print(type(k))
                if type(k)!= int and type(k)!=float:
                    return False
    length = len(x[0])
    # print(length)
    for i in range(1,len(x)):
        if len(x[i]) == length:
            return True
        else:
            return False
        
```

```python
# YOUR CODE HERE
def reverse_tuple(tup):
    if type(tup) != tuple:
         raise TypeError("Input is not a tuple!")
    else: 
        out = tuple(reversed(tup))
    return out
```

```python
assert type(reverse_tuple((4,5,6))) == tuple
```

```python
assert reverse_tuple((3,4,1)) == (1,4,3)
assert reverse_tuple((4,1)) == (1,4)
```
