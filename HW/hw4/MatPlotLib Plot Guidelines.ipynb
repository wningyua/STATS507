```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
norm01 = np.random.normal(size=1000)
norm34 = np.random.normal(loc=3, scale=2, size=1000)
```

<b><font color='red' size = 5>Wrong Way</font></b>

```python
plt.scatter(norm01, norm34)
plt.show()
```

<b><font color='green' size = 5>Right Way</font></b>

```python
fig = plt.figure()
ax = fig.gca()
ax.scatter(norm01, norm34)
plt.show()
```

```python
assert ax is not None
```
