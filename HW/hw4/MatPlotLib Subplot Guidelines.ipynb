```python
import numpy as np
import matplotlib.pyplot as plt
```

Say I wanted to plot some historgrams of samples from different normal distributions.

```python
norm01 = np.random.normal(size=1000)
norm04 = np.random.normal(scale=2, size=1000)
norm31 = np.random.normal(loc=3, size=1000)
norm34 = np.random.normal(loc=3, scale=2, size=1000)
```

<b><font color='red' size = 5>Wrong Way</font></b>

```python
# Generate Subplots
plt.subplots(4,1, figsize=(15,9))

# Access Each Subplot
plt.subplot(411)
plt.hist(norm01)

plt.subplot(412)
plt.hist(norm04)

plt.subplot(413)
plt.hist(norm31)

plt.subplot(414)
plt.hist(norm34)
plt.tight_layout()
```

```python
assert ax.data is not None
```

<b><font color='green' size = 5>Right Way</font></b>

```python
fig, ax = plt.subplots(4,1, figsize=(15,9))

ax[0].hist(norm01)
ax[1].hist(norm04)
ax[2].hist(norm31)
ax[3].hist(norm34)
plt.show()
```

```python
assert ax.data is not None
```
